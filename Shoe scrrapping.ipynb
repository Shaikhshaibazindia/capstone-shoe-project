```python
import selenium             # library used to work with selenium
import pandas as pd           
import numpy as np
import warnings              #To ignore any sort of warnings 
warnings.filterwarnings('ignore')
from selenium import webdriver     #importing webdriver module from selenium to open automated chrome window
from selenium.webdriver.common.by import By  #importing inbulit class By
import time                        #use to stop search engine for few seconds
from selenium.webdriver.common.action_chains import ActionChains

#for handling the exceptions while scrapping
from selenium.common.exceptions import NoSuchElementException,ElementNotInteractableException,TimeoutException
from selenium.webdriver.support import expected_conditions as EC
```

```python
#Opening the nike website in automated driver page 
driver=webdriver.Chrome()
driver.get('https://www.nike.com')
```

```python
#clicking on the mens category

Men=driver.find_element(By.XPATH,"/html/body/div[3]/div/div[3]/header/div[1]/div[2]/nav/div[2]/div/ul/li[2]/a")
Men.click()

Shoes=driver.find_element(By.CLASS_NAME,"JSftBPEZ")
Shoes.click()
```

```python
for _ in range(30):
    driver.execute_script("window.scroll(0,500)")
    
#Scraping the Shoe Name, Category ,NO of colors and Price of Men Catagory
Shoe_Name=[]
Category=[]
No_color=[]
price=[]


time.sleep(5)

shoe=driver.find_elements(By.XPATH,'//a[@class="product-card__link-overlay"]')
for i in shoe:
    Shoe_Name.append(i.text)
    
category=driver.find_elements(By.XPATH,'//div[@class="product-card__subtitle"]')
for i in category:
    Category.append(i.text)

    
color2=driver.find_elements(By.XPATH,'//*[@class="product-card__count-item"]/button/div')
for i in color2:
    No_color.append(i.text.split()[0])   
    
price_1=driver.find_elements(By.XPATH,'//*[@class="product-card__animation_wrapper"]/div/div/div/div')
for i in price_1: 
    a=i.text.replace("MRP :","").replace("₹", "")
    price.append(a)

```

```python
print(len(Shoe_Name),len(Category),len(No_color),len(price))
```

```python
men_Table1=pd.DataFrame({"ShoeName":Shoe_Name,"Category":Category,"No of Colors":No_color,"Price":price})
men_Table1
```

```python
driver.quit()
```

```python
#Opening the addidas website in automated driver page 
driver=webdriver.Chrome()
driver.get('https://www.adidas.com')
```

```python
#By clicking on the men category and then clicking on the all footwears

men_tags=driver.find_element(By.XPATH,'//*[@id="__next"]/div[2]/div[2]/div/header/div[2]/div/ul/li[1]/a')

#ActionChains instance
hover=ActionChains(driver)
# Move the cursor to the hover element
hover.move_to_element(men_tags).perform()

#clicking on the mens footwear

shoes_tags=driver.find_element(By.XPATH,'/html/body/div[1]/div[2]/div[2]/div/header/div[2]/div/ul/li[1]/div/ul[1]/li[2]/div/a/div')
shoes_tags.click()
```

```python
# for scrapping the Shoe name ,Category, no of colors, price

Shoe_Name_A=[]
Category_A=[]
No_color_A=[]
price_A=[]

Shoe_a=driver.find_elements(By.XPATH,'//*[@class="glass-product-card__title"]')
for i in Shoe_a:
     Shoe_Name_A.append(i.text)
        
cat_a=driver.find_elements(By.XPATH,'//*[@class="glass-product-card__category"]')
for i in cat_a:
    Category_A.append(i.text)

    
try:
    no_color_a=driver.find_elements(By.XPATH,'//*[@class="dark-grey___6ysQv"]')
    for i in no_color_a:
        No_color_A.append(i.text.split()[0])
except NoSuchElementException:
    No_color_A.append("-")

price_1=driver.find_elements(By.XPATH,'//*[@class="gl-price gl-price--horizontal notranslate gl-price--inline___3nMlh"]')
for i in price_1:
    m=i.text.replace("₹", "")
    price_A.append(m.split("\n")[0])
```

```python
print(len(Shoe_Name_A),len(Category_A),len(No_color_A),len(price_A))
```

```python
while len(No_color_A)<len(Shoe_Name_A):
    
    
    No_color_A.append("-")
    
while len(price_A)<len(Shoe_Name_A):
    
    
    price_A.append("-")
```

```python
print(len(Shoe_Name_A),len(Category_A),len(No_color_A),len(price_A))
```

```python
men_Table1_A=pd.DataFrame({"ShoeName":Shoe_Name_A,"Category":Category_A,"No of Colors":No_color_A,"Price":price_A})
men_Table1_A
```
